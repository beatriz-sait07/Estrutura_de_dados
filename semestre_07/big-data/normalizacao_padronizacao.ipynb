```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('vinho.csv')
df
```

```python
#criando objt
scl_n = MinMaxScaler()

#normalizando os dados
df_normalizado = scl_n.fit_transform(df)

#criando o df normalizados
df_normalizado = pd.DataFrame(df_normalizado, columns=df.columns)
df_normalizado
```

```python
scl_p = StandardScaler()

#padronizando os dados
df_pad = scl_p.fit_transform(df)

df_pad = pd.DataFrame(df_pad, columns=df.columns)
df_pad
```
